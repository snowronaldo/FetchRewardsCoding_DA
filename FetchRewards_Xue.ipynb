# Correct Receipts

```python
receipts = pd.read_json('receipts.json',  lines=True)

receipts.head()
```

```python
# id column
# for every row in column '_id', make the value as value of key->'$oid'
receipts['_id'] = receipts['_id'].apply(lambda x: x['$oid'])
receipts.head()
```

```python
# date columns
#
for i in receipts.columns:
    if re.findall('date', i.lower())!=[]:
        receipts[i] = receipts[i].apply(lambda x: datetime.fromtimestamp(x['$date'] // 1000) if str(x) != str(np.nan) else x)
receipts.head()
```

```python
# check if bonusPointsEarned == pointsEarned
receipts['bonusPointsEarned'].equals(receipts['pointsEarned'])
```

## Deal with rewardsReceiptItemList Column

```python
# get all keys in rewardsReceiptItemList column

RRIL_keys = []

for i in receipts[~receipts.rewardsReceiptItemList.isnull()].rewardsReceiptItemList:
    RRIL_keys.extend(i[0].keys())
    
# delete duplicated keys    
RRIL_keys = list(set(RRIL_keys))
print(RRIL_keys)
```

```python
# create a new table for rewardsReceiptItemList
RRIL = pd.DataFrame(columns=['_id'] + RRIL_keys)

for i in range(receipts.shape[0]):
    if isinstance(receipts.rewardsReceiptItemList[i], list):
        receipts.rewardsReceiptItemList[i][0]['_id'] = receipts._id[i]
```

```python
for i in range(receipts.shape[0]):
    if isinstance(receipts.rewardsReceiptItemList[i], list):
        tempt = pd.DataFrame(receipts.rewardsReceiptItemList[i][0], index=[0])
        RRIL = RRIL.append(tempt, sort=False).reset_index(drop=True)
```

```python
RRIL.head()
```

```python
# check any duplicates
print(RRIL[RRIL.duplicated(subset = '_id', keep = False)])

#check any null value in '_id'
print(RRIL['_id'].isnull().any())
```

```python
# delete rewardsReceiptItemList column from receipts table
del receipts['rewardsReceiptItemList']
receipts.head()
```

## 1. Top 5 brands by receipts scanned for most recent month?

```python
# Calculate top5 brands for the latest month
engine.execute("WITH etc AS (\
                    SELECT \
                        strftime('%Y%m',datetime(re.dateScanned)) AS dateScanned, \
                        RRIL.brandCode, COUNT(re._id) AS cnt, \
                        ROW_NUMBER() OVER (PARTITION BY strftime('%Y%m',datetime(re.dateScanned)) ORDER BY COUNT(re._id) DESC) AS receipts_rnk, \
                        DENSE_RANK() OVER (ORDER BY strftime('%Y%m',datetime(re.dateScanned)) DESC) AS dateScanned_rnk \
                    FROM receipts re \
                    LEFT JOIN RRIL \
                    ON re._id=RRIL._id \
                    WHERE RRIL.brandCode IS NOT NULL \
                    GROUP BY \
                        strftime('%Y%m',datetime(re.dateScanned)), \
                        RRIL.brandCode \
                    ORDER BY \
                       strftime('%Y%m',datetime(re.dateScanned)) DESC, \
                       receipts_rnk) \
               SELECT *\
               FROM etc \
               WHERE receipts_rnk <= 5 AND dateScanned_rnk = 1\
               ORDER BY cnt DESC;").fetchall()

```

For the most recent month, only two brands have records, so only 2 bands in Top5: 'BRAND' and 'MISSION'.

## 2. How does the ranking of the top 5 brands by receipts scanned for the recent month compare to the ranking for the previous month?

```python
# Calculate top5 brands for the 2nd latest month
engine.execute("WITH etc AS (\
                    SELECT \
                        strftime('%Y%m',datetime(re.dateScanned)) AS dateScanned, \
                        RRIL.brandCode, COUNT(re._id) AS cnt, \
                        ROW_NUMBER() OVER (PARTITION BY strftime('%Y%m',datetime(re.dateScanned)) ORDER BY COUNT(re._id) DESC) AS receipts_rnk, \
                        DENSE_RANK() OVER (ORDER BY strftime('%Y%m',datetime(re.dateScanned)) DESC) AS dateScanned_rnk \
                    FROM receipts re \
                    LEFT JOIN RRIL \
                    ON re._id=RRIL._id \
                    WHERE RRIL.brandCode IS NOT NULL \
                    GROUP BY \
                        strftime('%Y%m',datetime(re.dateScanned)), \
                        RRIL.brandCode \
                    ORDER BY \
                       strftime('%Y%m',datetime(re.dateScanned)) DESC, \
                       receipts_rnk) \
               SELECT *\
               FROM etc \
               WHERE receipts_rnk <= 5 AND dateScanned_rnk = 2\
               ORDER BY cnt DESC;").fetchall()
```

For the previous month, Top5 brands are:
    1.'BRAND' 2.'HY-VEE' 3.'BEN AND JERRYS' 4.'WINGSTOP' 5.'MISSION'
    

## 3. When considering average spend from receipts with 'rewardsReceiptStatus' of 'Accepted' or 'Rejected', which is greater?

```python
engine.execute("SELECT\
                    rewardsReceiptStatus \
                FROM (SELECT \
                            DISTINCT rewardsReceiptStatus, \
                            RANK() OVER (PARTITION BY rewardsReceiptStatus ORDER BY AVG(totalSpent) DESC) AS rnk \
                      FROM receipts \
                      WHERE rewardsReceiptStatus IN ('FINISHED','REJECTED')) \
                WHERE rnk = 1;").fetchall()
```

Receipts with status of finished are greater.

## 4. When considering total number of items purchased from receipts with 'rewardsReceiptStatus’ of ‘Accepted’ or ‘Rejected’, which is greater?

```python
engine.execute("SELECT\
                    rewardsReceiptStatus \
                FROM (SELECT \
                            DISTINCT rewardsReceiptStatus, \
                            RANK() OVER (PARTITION BY rewardsReceiptStatus ORDER BY SUM(purchasedItemCount) DESC) AS rnk \
                      FROM receipts \
                      WHERE rewardsReceiptStatus IN ('FINISHED','REJECTED')) \
                WHERE rnk = 1;").fetchall()
```

Still receipts with status of finished are greater.

## 5. Which Brand has the most spend among users who were created within the past 6 months?

```python
engine.execute("SELECT RRIL.brandCode, SUM(r.totalSpent)\
                FROM RRIL \
                LEFT JOIN receipts r   \
                ON RRIL._id = r._id \
                LEFT JOIN users u\
                ON r.userId = u._id \
                WHERE u._id in (SELECT _id \
                                FROM USERS \
                                WHERE date(datetime(createdDate/1000,'unixepoch')) \
                                > (select date(max(datetime(createdDate/1000,'unixepoch')), '-6 months') FROM USERS )\
                                ) \
                      AND RRIL.brandCode IS NOT NULL \
                GROUP BY RRIL.brandCode \
                ORDER BY SUM(r.totalSpent) DESC LIMIT 1").fetchall()
```

## 6. Which brand has the most transactions among users who were created within the past 6 months?

```python
engine.execute("SELECT RRIL.brandCode, COUNT(*)\
                FROM RRIL \
                LEFT JOIN receipts r   \
                ON RRIL._id = r._id \
                LEFT JOIN users u\
                ON r.userId = u._id \
                WHERE u._id in (SELECT _id \
                                FROM USERS \
                                WHERE date(datetime(createdDate/1000,'unixepoch')) \
                                > (select date(max(datetime(createdDate/1000,'unixepoch')), '-6 months') FROM USERS )\
                                ) \
                      AND RRIL.brandCode IS NOT NULL \
                GROUP BY RRIL.brandCode \
                ORDER BY COUNT(*) DESC LIMIT 1").fetchall()
```
